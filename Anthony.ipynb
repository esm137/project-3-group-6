```python
# Display Top 20 Artists

top20 = RankingsData.groupby("Artist")["Streams"].sum().sort_values(ascending=False).head(20)

```

```python
# display(top20) Artist by Sreams

top20.head()
```

```python
# Display Barchart
top20.plot(kind='bar')
plt.title(" Sportify Stream by Atrists")
plt.xlabel("Artist")
plt.ylabel("Number of Streams")
```
